Fazer um progama que o usuario irá entrar com 04 notas, vamos calcular a média e mostrar o usuário

```python
nota1 =  float(input("Digite o número:"))
nota2 =  float(input("Digite o número:"))
nota3 =  float(input("Digite o número:"))
nota4 =  float(input("Digite o número:"))

#Processamento
Média = (nota1 + nota2 + nota3 + nota4) / 4

#Saida

print (f"A média é {Média}")
```

```python
numero1 = int(input("Digite um número:"))
numero2 = int(input("Digite um número:"))
#Processamento e saída
if numero1 > numero2:
    print(f"{numero1} é maior que {numero2}")
elif numero1 == numero2:
    print(f"O número{numero1} é igual ao número{numero2}")
else:
    print(f"{numero1} é menor que {numero2}.")
```
